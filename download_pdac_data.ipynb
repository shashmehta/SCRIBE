## Prepare Data for Classifier

### Subtask:
Extract the gene expression matrix (`adata.X`) as features and the 'CONDITION' column from `adata.obs` as target labels. Convert the sparse expression matrix to a dense array and encode the string labels into numerical format suitable for scikit-learn. ## Prepare Data for Classifier

### Subtask:
Encode the categorical 'CONDITION' labels into numerical format using `LabelEncoder` and verify the transformation.


**Reasoning**:
I will import `LabelEncoder`, initialize it, fit and transform the `y` variable to numerical format, print the inverse transformation to verify the labels, display the first few encoded labels, and finally print the shapes of `X` and `y_encoded` to confirm their dimensions.



```python
from sklearn.preprocessing import LabelEncoder

# 3. Initialize LabelEncoder
le = LabelEncoder()

# 4. Fit and transform the 'CONDITION' labels
y_encoded = le.fit_transform(y)

print("Labels encoded.")

# 5. Print the original labels corresponding to the encoded numerical values
print("\nOriginal labels corresponding to encoded values:")
for i, label in enumerate(le.classes_):
    print(f"{i}: {label}")

# 6. Print the first few encoded labels
print("\nFirst 5 encoded labels:")
print(y_encoded[:5])

# 7. Print the shape of X and y_encoded
print(f"\nShape of features (X): {X.shape}")
print(f"Shape of encoded target (y_encoded): {y_encoded.shape}")
```

## Split Data into Training and Testing Sets

### Subtask:
Divide the prepared feature matrix and target labels into training and testing datasets using `train_test_split` from scikit-learn, ensuring a reproducible split.


**Reasoning**:
I need to import `train_test_split` from `sklearn.model_selection` to split the feature matrix `X` and the encoded target labels `y_encoded` into training and testing sets. This is a crucial step for preparing the data for model training and evaluation.



```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

print("Data split into training and testing sets.")

# Print the shapes of the resulting sets to verify the split
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

## Calculate Average Gene Expression in Normal Cells

### Subtask:
Filter the AnnData object to include only 'N' (Normal) cells from the 'CONDITION' column. Then, calculate the average expression of each gene across these normal cells. Store these average expressions in a pandas Series.


**Reasoning**:
First, I will filter the `adata` object to include only 'N' (Normal) cells based on the 'CONDITION' column, which is a necessary initial step to calculate average gene expression specifically for normal cells.



```python
normal_cells = adata[adata.obs['CONDITION'] == 'N']
print("Filtered AnnData object for normal cells created.")
print(f"Shape of normal_cells: {normal_cells.shape}")
```

**Reasoning**:
Now that the normal cells have been filtered, I will access their gene expression matrix, convert it to a dense array if sparse, calculate the mean expression for each gene, and store these averages in a pandas Series using gene names as the index.



```python
import scipy.sparse

# Access the gene expression matrix for normal cells
# Convert to dense array if sparse before calculating mean
if isinstance(normal_cells.X, scipy.sparse.spmatrix):
    normal_cells_X_dense = normal_cells.X.toarray()
else:
    normal_cells_X_dense = normal_cells.X

# Calculate the mean expression for each gene across these normal cells
normal_avg_expression = pd.Series(normal_cells_X_dense.mean(axis=0), index=adata.var_names)

print("Average gene expression for normal cells calculated and stored in 'normal_avg_expression'.")
print("First 5 average expressions:")
print(normal_avg_expression.head())
```

**Reasoning**:
Now, I will filter the `adata` object to include only 'T' (Tumor) cells based on the 'CONDITION' column to prepare for calculating the average gene expression specifically for tumor cells.



```python
tumor_cells = adata[adata.obs['CONDITION'] == 'T']
print("Filtered AnnData object for tumor cells created.")
print(f"Shape of tumor_cells: {tumor_cells.shape}")
```

**Reasoning**:
Now that the tumor cells have been filtered, I will access their gene expression matrix, convert it to a dense array if sparse, calculate the mean expression for each gene, and store these averages in a pandas Series using gene names as the index.



```python
import scipy.sparse

# Access the gene expression matrix for tumor cells
# Convert to dense array if sparse before calculating mean
if isinstance(tumor_cells.X, scipy.sparse.spmatrix):
    tumor_cells_X_dense = tumor_cells.X.toarray()
else:
    tumor_cells_X_dense = tumor_cells.X

# Calculate the mean expression for each gene across these tumor cells
tumor_avg_expression = pd.Series(tumor_cells_X_dense.mean(axis=0), index=adata.var_names)

print("Average gene expression for tumor cells calculated and stored in 'tumor_avg_expression'.")
print("First 5 average expressions:")
print(tumor_avg_expression.head())
```

**Reasoning**:
Now that I have the average gene expression for both normal and tumor cells, I will compute the ratio of normal to tumor cell expression for each gene. I need to handle potential division by zero by replacing any zero tumor expression values with a small epsilon to avoid `inf` or `NaN` values, ensuring a valid ratio calculation.



```python
import numpy as np

# Ensure both Series have the same index (gene names)
# This should already be the case since both were indexed by adata.var_names

# Handle potential division by zero: replace 0 in tumor_avg_expression with a small epsilon
# This assumes expression values are non-negative or we want to avoid issues with negative values becoming very large ratios
# A small positive epsilon is appropriate for gene expression data where 0 typically means no expression.
epsilon = 1e-6 # A very small number
tumor_avg_expression_safe = tumor_avg_expression.replace(0, epsilon)
normal_avg_expression_safe = normal_avg_expression.replace(0, epsilon)


# Calculate the ratio of normal to tumor cell expression for each gene
gene_expression_ratio = normal_avg_expression_safe / tumor_avg_expression_safe

print("Ratio of normal to tumor cell expression calculated.")
print("First 5 gene expression ratios:")
print(gene_expression_ratio.head())
```
